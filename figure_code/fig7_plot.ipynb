```python
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
plt.rc('font', size=20)        # 기본 폰트 크기
plt.rc('axes', labelsize=20)   # x,y축 label 폰트 크기
plt.rc('xtick', labelsize=15)  # x축 눈금 폰트 크기 
plt.rc('ytick', labelsize=15)  # y축 눈금 폰트 크기
plt.rc('legend', fontsize=15)  # 범례 폰트 크기
plt.rc('figure', titlesize=100)
```

```python
with open('noise_loss_list_withoutR.pickle', 'rb') as f:
    noise_loss_list = pickle.load(f)
with open('clean_loss_list_withoutR.pickle', 'rb') as f:
    clean_loss_list = pickle.load(f)


plt.figure(figsize=(8, 6))

# 히스토그램 (밀도로 표시)
plt.hist(clean_loss_list, bins=150, density=True, alpha=0.6, color='b', label='Clean')
sns.kdeplot(clean_loss_list, color='blue', label='KDE')

plt.hist(noise_loss_list, bins=150, density=True, alpha=0.6, color='r', label='Noise')
sns.kdeplot(noise_loss_list, color='red', label='KDE')

# 그래프 제목 및 레이블 추가
plt.xlim([0,3])
plt.title('CE')
plt.xlabel('Loss')
plt.ylabel('parobability density')
plt.legend()

# 그래프 보여주기
plt.show()
```

```python
with open('noise_loss_list_withR.pickle', 'rb') as f:
    noise_loss_list = pickle.load(f)
with open('clean_loss_list_withR.pickle', 'rb') as f:
    clean_loss_list = pickle.load(f)


plt.figure(figsize=(8, 6))

# 히스토그램 (밀도로 표시)
plt.hist(clean_loss_list, bins=150, density=True, alpha=0.6, color='b', label='Clean')
sns.kdeplot(clean_loss_list, color='blue', label='KDE')

plt.hist(noise_loss_list, bins=150, density=True, alpha=0.6, color='r', label='Noise')
sns.kdeplot(noise_loss_list, color='red', label='KDE')

# 그래프 제목 및 레이블 추가
plt.xlim([0,3])
plt.title('CE+R')
plt.xlabel('Loss')
plt.ylabel('parobability density')
plt.legend()

# 그래프 보여주기
plt.show()
```
