```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


def get_only_acc(filename) : # log에서 accuracy만 추출

    with open(filename) as f :
        text = f.read()

    res = []
    for t in text.split('Accuracy: ')[1:] :
        res.append(float(t[:5]))

    return res


def get_only_f1(filename) : # log에서 F1 score만 추출

    with open(filename) as f :
        text = f.read()

    f1 = [0 for _ in range(10)]

    res = []
    for t in text.split('f-score: ')[1:] :
        res.append(float(t[:5]))
    f1 = f1 + res
    return f1


res_k2 = get_only_acc('../log/base/cifar10_instance_0.4_k=2_test.txt')  # 예) instance 40% / bank size T=2 log
res_k3 = get_only_acc('../log/base/cifar10_instance_0.4_k=3_test.txt')
res_k4 = get_only_acc('../log/base/cifar10_instance_0.4_k=4_test.txt')
res_k5 = get_only_acc('../log/base/cifar10_instance_0.4_k=5_test.txt')

plt.figure(figsize=(5,4))
plt.xlim(10,75)
plt.ylim(70,100)
plt.plot(res_k2, color='#FF3D3D', label='T=2')
plt.plot(res_k3, color='#3A9EFF', label='T=3')
plt.plot(res_k4, color='#B77CED', label='T=4')
plt.plot(res_k5, color='#209090', label='T=5')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Test Acc.(%)')
plt.title('Inst 40%')
plt.show()
```

```python
f1_k2 = get_only_f1('../log/base/cifar10_instance_0.4_k=2_test.txt')  # 예) instance 20% / bank size T=2 log
f1_k3 = get_only_f1('../log/base/cifar10_instance_0.4_k=3_test.txt')
f1_k4 = get_only_f1('../log/base/cifar10_instance_0.4_k=4_test.txt')
f1_k5 = get_only_f1('../log/base/cifar10_instance_0.4_k=5_test.txt')

plt.figure(figsize=(5,4))
plt.xlim(10,75)
plt.ylim(0.86,1)
plt.plot(f1_k2, '-.', color='#FF3D3D', label='T=2')
plt.plot(f1_k3, '-.', color='#3A9EFF', label='T=3')
plt.plot(f1_k4, '-.', color='#B77CED', label='T=4')
plt.plot(f1_k5, '-.', color='#209090', label='T=5')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('F-score')
plt.title('Inst 40%')
plt.show()
```
